```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import Perceptron
```

```python
# Erzeugen von "Punktehaufen" die linear trennbar sind
X,y = datasets.make_blobs(n_samples=100,n_features=2,centers=2,random_state=3)
```

```python
# Klassifikation
# Aufbau eines Rasters um auszuwerten und zu zeichnen
s = 0.02 # Schrittweite
x_min, x_max = X[:,0].min()- 1,X[:,0].max() + 1
y_min, y_max = X[:,1].min()- 1,X[:,1].max() + 1
xx,yy = np.meshgrid(np.arange(x_min,x_max,s),np.arange(y_min,y_max,s))
```

```python
Perceptron = Perceptron(random_state = 43, max_iter=1000)
Perceptron.fit(X,y)
```

```python
# ravel() erzeugt 1-D-Array
# np.c_ erzeugt Punktpärchen Arrays für jeden Rasterpunkt
Prediction = Perceptron.predict(np.c_[xx.ravel(),yy.ravel()])
```

```python
plt.plot(X[:,0][y==0],X[:,1][y==0],'b^') # blaue Dreiecke
plt.plot(X[:,0][y==1],X[:,1][y==1],'ys') # gelbe Quadrate
# Umwandlung von 1-D Arrays in die Rasterdimension
Prediction = Prediction.reshape(xx.shape)
plt.contourf(xx,yy,Prediction,cmap=plt.cm.Paired)
plt.show()
```
